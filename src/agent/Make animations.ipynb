```julia
using Pkg;Pkg.activate("../..");Pkg.instantiate()
using Plots, OhMyREPL, JLD
using Statistics:mean,cov
pyplot()

experiment = JLD.load("batch_experiment.jld")
traj = experiment["traj"]

anim = @animate for t in 1:80
    scatter(traj[:,:,t][:,1],traj[:,:,t][:,2],ylim=(0,1),xlim=(0,1),title = "Timepoint: " *repr(t));
end

gif(anim,"batch_run.gif",fps=2)
```
